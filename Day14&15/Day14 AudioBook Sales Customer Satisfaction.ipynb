```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
#Load data from a text file
raw_data=np.loadtxt('Audiobooks_data.csv',delimiter=',')
print(raw_data)
print(raw_data.shape)
```

```python
#Semua Rows, Exclude row index 1 dan index -1 (Paling depan dan paling belakang)
unscaled_input=raw_data[:,1:-1]
print(unscaled_input)
print(unscaled_input.shape)
```

```python
#Include semua rows, hanya index column terakhir yang di include
target=raw_data[:,-1]
print(target)
print(target.shape)
```

```python
#Jumlahin semua total target yg ada di Data
one=int(np.sum(target))
zero=0
indices_remove=[]

print(one)
```

```python
#Target shape [0] is the length of vector
target.shape[0]
```

```python
for i in range(target.shape[0]):
    if target[i]==0:
        zero +=1
        if zero>one:
            indices_remove.append(i)
```

```python
indices_remove
```

np.delete(array,obj to delete,axis) is a method that deletes an object along an axis

```python
new_unscaled=np.delete(unscaled_input,indices_remove,axis=0)
print(new_unscaled)
print(new_unscaled.shape)
```

```python
new_target=np.delete(target,indices_remove,axis=0)
print(new_target)
print(new_target.shape)
```

# Standardize Input & Target

preprocessing.scale(x) is a method that standardize arrays

```python
scaled_inputs=preprocessing.scale(new_unscaled)
print(scaled_inputs)
print(scaled_inputs.shape)
```

```python
scaled_inputs.shape[0]
```

np.arange([start],stop) is a method that returns evenly spaced values within a given interval

```python
#INDICES ITU NOMER/ID/ROWS
shuffled_indices=np.arange(scaled_inputs.shape[0])
shuffled_indices
```

np.random.shuffle(X) is a method that shuffles the numbers in a given sequence

```python
np.random.shuffle(shuffled_indices)
```

```python
shuffled_inputs=scaled_inputs[shuffled_indices]
shuffled_targets=new_target[shuffled_indices]
```

```python
print(shuffled_inputs)
print(shuffled_inputs.shape)
```

```python
print(shuffled_targets)
print(shuffled_targets.shape)
```

# Split Data

```python
sample=shuffled_inputs.shape[0]
train=int(0.8*sample)
validation=int(0.1*sample)
test=sample-train-validation

print(train)
print(validation)
print(test)
```

```python
#[:3579] -> 3579 elemen pertama dalam list
train_input=shuffled_inputs[:train]
train_target=shuffled_targets[:train]

print(train_input.shape)
print(train_target.shape)
```

```python
#[3579:4026]-> Elemen ke 3580 sampe ke 4026 jadi VALIDATION DATASET
validation_input=shuffled_inputs[train:train+validation]
validation_target=shuffled_targets[train:train+validation]

print(validation_input.shape)
print(validation_target.shape)
```

```python
#[4026:]-> Semua elemen lain selain 4026 di awal
test_input=shuffled_inputs[train+validation:]
test_target=shuffled_targets[train+validation:]

print(test_input.shape)
print(test_target.shape)
```

```python
#TARGET data vs INPUT data
#Target data di train, validation, test sama" di 49-52%
#Berarti data balanced
print(np.sum(train_target)/train)
print(np.sum(validation_target)/validation)
print(np.sum(test_target)/test)
```

# Save Dataset

```python
np.savez('Audiobook_train',inputs=train_input,targets=train_target)
np.savez('Audiobook_validation',inputs=validation_input,targets=validation_target)
np.savez('Audiobook_test',inputs=test_input,targets=test_target)
```
