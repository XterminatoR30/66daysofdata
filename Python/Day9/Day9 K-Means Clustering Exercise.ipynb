```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Geographical Segmentation

```python
data=pd.read_csv('Countries-exercise.csv')
data
```

```python
x=data.iloc[:,1:3]
x
```

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(7)
km.fit(x)
```

```python
clusters=km.fit_predict(x)
clusters
```

```python
data_cluster=data
data_cluster['Cluster Prediction']=clusters
data_cluster
```

```python
plt.scatter(data['Longitude'],data['Latitude'],c=data_cluster['Cluster Prediction'],cmap='rainbow')
plt.xlim(-180,180)
plt.xlabel('Longitude',fontsize=16)
plt.ylim(-90,90)
plt.ylabel('Latitude',fontsize=16)
plt.show()
```

```python
data2=pd.read_csv('Categorical.csv')
data2
```

```python
data2['continent'].unique()
```

```python
continent_data=data2
continent_data['continent']=continent_data['continent'].map({'North America':0,'Asia':1,'Africa':2,'Europe':3,'South America':4,'Oceania':5,'Antarctica':6,'Seven seas (open ocean)':7})
continent_data
```

```python
continent_data.isnull().sum()
```

```python
pd.options.display.max_rows=241
x2=continent_data.iloc[:,3:4]
x2
```

```python
cont_km=KMeans(7)
print(cont_km.fit(x2))
cont=cont_km.fit_predict(x2)
print(cont)
```

```python
new_cont=continent_data
new_cont['Clusters']=cont
new_cont
```

```python
plt.scatter(new_cont['Longitude'],new_cont['Latitude'],c=new_cont['Clusters'],cmap='rainbow')
plt.xlim(-180,180)
plt.xlabel('Longitude',fontsize=16)
plt.ylim(-90,90)
plt.ylabel('Latitude',fontsize=16)
plt.show
```

# WCSS (Within Cluster Sum of Squares)

```python
cont_km.inertia_
```

list.append(x) appends a new item with value of x to the end of a list

```python
wcss=[]
for i in range(1,8):
    cont_km=KMeans(i)
    cont_km.fit(x2)
    wcss_iter= cont_km.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
num_cluster=range(1,8)
plt.plot(num_cluster,wcss)
plt.title('The Elbow Method')
plt.xlabel('Num of Clusters')
plt.ylabel('WCSS (Within Cluster Sum of Squares)')
#Based on this elbow method visualization, the optimum number of clusters is 5
#As after 5 clusters, there are no significant decrease in WCSS Value
```

# Species Segmentation

```python
data3=pd.read_csv('iris_dataset.csv')
data3
```

```python
data3.describe(include='all')
```

```python
plt.scatter(data3['sepal_length'],data3['sepal_width'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
x3=data3.iloc[:,0:2]
x3
```

```python
from sklearn import preprocessing
x3_scaled=preprocessing.scale(x3)
x3_scaled
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,10):
    kmeans3=KMeans(i)
    kmeans3.fit(x3_scaled)
    wcss.append(kmeans3.inertia_)
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans4=KMeans(3)
km_fit=kmeans4.fit(x3_scaled)
print(km_fit)
pred=km_fit.fit_predict(x3_scaled)
print(pred)
```

```python
newdata3=x3.copy()
newdata3['Cluster Prediction']=pred
newdata3
```

```python
plt.scatter(newdata3['sepal_length'],newdata3['sepal_width'],c=newdata3['Cluster Prediction'],cmap='rainbow')
plt.title('Sepal Length vs Sepal Width K-Means Clustering')
plt.xlabel('Sepal Length',fontsize=16)
plt.ylabel('Sepal Width',fontsize=16)
```

```python
y3=data3.iloc[:,2:4]
y3
```

```python
y3_scaled=preprocessing.scale(y3)
y3_scaled
```

```python
kmeans5=KMeans(3)
km_fit5=kmeans5.fit(y3_scaled)
print(km_fit5)
pred5=km_fit5.fit_predict(y3_scaled)
print(pred5)
```

```python
ydata=y3.copy()
ydata['Cluster Prediction']=pred5
ydata
```

```python
plt.scatter(ydata['petal_length'],ydata['petal_width'],c=ydata['Cluster Prediction'],cmap='rainbow')
plt.title('Petal Length vs Petal Width K-Means Clustering')
plt.xlabel('Petal Length',fontsize=16)
plt.ylabel('Petal Width',fontsize=16)
```
