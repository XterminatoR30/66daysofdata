```python
import numpy as np
import matplotlib.pyplot as plt
#Buat 3D Graphs
from mpl_toolkits.mplot3d import Axes3D
```

np.random.uniform(low,high,size) draws a random value from interval (low,high) where each number has an equal chance to be selected

np.column_stack(appropriate tuples) takes a sequence of 1D arrays and stacks them into single 2D array

```python
observations=1000
x=np.random.uniform(low=10,high=10,size=(observations,1))
z=np.random.uniform(-10,10,(observations,1))
print(x)
print(z)
inputs= np.column_stack((x,z))
print(inputs)
print(inputs.shape)
```

```python
noise=np.random.uniform(-1,1,(observations,1))
print(noise)
targets=2*x-3*z+5+noise
print(targets)
print(targets.shape)
```

```python
#Learning Rate detetrmines how fast a model can train
lr=0.02
init_range=0.1
weights=np.random.uniform(low=-init_range,high=init_range,size=(2,1))
bias=np.random.uniform(low=-init_range,high=init_range,size=1)
print(weights)
print(bias)
```

# Train Model

Calculate Outputs

Compare Outputs to Targets through the Loss

Print Loss

Adjust Weights and Biases

```python
#np.sum(a) method to sum all values in the array
for i in range(100):
    #y=xw+b (Linear Model Equation)
    outputs= np.dot(inputs,weights) +bias
    
    #Delta: differences between outputs
    delta= outputs-targets
    
    #L2-Norm Regression Loss Function Equation
    loss=np.sum(delta**2)/2/observations
    
    #Print loss 100 times
    print('Loss:',loss)
    delta_scale=delta/observations
    
    #Update Weights and Biases for Gradient Descent Optimization
    weights=weights-lr * np.dot(inputs.T,delta_scale)
    bias= bias-lr*np.sum(delta_scale)
```

```python
print(weights,bias)
```

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
