```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import scipy
```

```python
data=pd.read_csv('real_estate_price_size.csv')
data
```

```python
data.describe()
```

```python
import statsmodels.api as sm
from statsmodels.api import add_constant
import seaborn as sns
sns.set()
```

```python
x1=data['size']
y=data['price']
plt.scatter(x1,y,c='blue')
plt.xlabel('Size',fontsize=20)
plt.ylabel('Price',fontsize=20)
plt.show()
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(x1,y)
yhat=223.1787*x1+101900
fig=plt.plot(x1,yhat,lw=4,c='red')
plt.xlabel('Size',fontsize=20)
plt.ylabel('Price',fontsize=20)
plt.show()
```

```python
data3=pd.read_csv('real_estate_price_size_year.csv')
data3
```

```python
data3.describe()
```

```python
y2=data3['price']
x2=data3[['size','year']]
x3=sm.add_constant(x2)
result2=sm.OLS(y2,x3).fit()
result2.summary()
#FINAL EQUATION
#yhat=227.7009*x1+2916.7853*x2-5772000

#CONCLUSION
#Both R Squared and Adj R Squared has very high value, therefore the model fits the data very well
#With the addition of new variable YEAR, it increases the Adj R squared value, therefore it's a significant variable for this model
#P value of all three variables are 0, which is <0.005, it further enhance the conclusion that these 3 values are very significant to each other
#Based on the DURBIN WATSON test, the score is 2.25 which is close to the value of 2. This means that the model has no autocorrelation assumption
```

```python
plt.scatter(data3['size'],y2)
yhat=227.7009*data3['size']+2916.7853*data3['year']-5772000
fig=plt.plot(data3['size'],yhat,lw=2,c='green')
plt.xlabel('Size',fontsize=20)
plt.ylabel('Price',fontsize=20)
plt.show()
```

```python
#ASSUMPTIONS
#1.Linearity
#2.No Endogeneity
#3.Normality and Homoscedacity
#4.No Autocorrelation
#5.No Mutlicollinearity
data4=pd.read_csv('1.03.+Dummies.csv')
data4
```

```python
#PANDAS
data5=data4.copy()
data5['Attendance']=data5['Attendance'].map({'Yes':1,'No':0})
```

```python
data5
```

```python
data5.describe()
```

```python
y4=data5['GPA']
x4=data5[['SAT','Attendance']]
x5=sm.add_constant(x4)
result4=sm.OLS(y4,x5).fit()
result4.summary()
#Yhat Equation
#yhat=0.0014*x1+0.2226*x2+0.6439

#Conclusion
#Based on the OLS model summary, it can be said that the model has a good fit with the data. As both R squared and Adj R squared value increases to >0.5
#Overall model also seems significant as F-Stats value is above 50
#All three variables are highly correlated as P value for SAT and attendance is below <0.005 threshold
#Since durbin watson value is still above 1, it can be said that the model has no autocorrelation observed
```

```python
plt.scatter(data4['SAT'],y4)
yhat_no=0.0014*data4['SAT']+0.6439
yhat_yes=0.0014*data4['SAT']+0.8665
fig=plt.plot(data4['SAT'],yhat_no,lw=2,c='red')
fig=plt.plot(data4['SAT'],yhat_yes,lw=2,c='green')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
plt.scatter(data4['SAT'],data4['GPA'])
yhat_no=0.0014*data4['SAT']+0.6439
yhat_yes=0.0014*data4['SAT']+0.8665
yhat=0.0017*data4['SAT']+0.275
fig=plt.plot(data4['SAT'],yhat_no,lw=2,c='red')
fig=plt.plot(data4['SAT'],yhat_yes,lw=2,c='green')
fig=plt.plot(data4['SAT'],yhat,lw=2,c='blue')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
