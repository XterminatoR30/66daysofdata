np.log(x) returns the natural logarithm of a number of array of numbers

```python
log_price=np.log(finaldata['Price'])
finaldata['Log Price']=log_price
finaldata
```

```python
#Dropping Price Column
finaldata2=finaldata.drop(['Price'],axis=1)
finaldata2
```

# Multicolinearity

```python
finaldata2.columns.values
```

VIF(Variance Inflation Factor)

```python
#Defining VIF Values
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables= finaldata2[['Mileage','Year','EngineV']]
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['Features']= variables.columns
vif
```

```python
#Dropping Year Column due to very high correlation with other variables (MULTICOLINEARITY) (VIF>10)
data_multicollinearity=finaldata2.drop(['Year'],axis=1)
data_multicollinearity
```

# Dummy Variables

pd.get_dummies(df,drop_first=True) spots all categorical variables and creates dummies automatically

```python
#If we have N categories for feature, we have to create N-1 dummies
data_dummy=pd.get_dummies(data_multicollinearity,drop_first=True)
data_dummy
```

```python
#Check VIF of features including the dummies
data_dummy.columns.values
```

```python
cols=['Log Price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
data_preprocessed=data_dummy[cols]
data_preprocessed
```

# Scaling and Transforming data

```python
targets=data_preprocessed['Log Price']
inputs=data_preprocessed.drop(['Log Price'],axis=1)
targets
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(inputs)
```

```python
scaled_inputs=scaler.transform(inputs)
scaled_inputs
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,test_size=0.2,random_state=365)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Create LOG Linear Regression

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
yhat=reg.predict(x_train)
print(yhat)
print(yhat.shape)
```

```python
plt.scatter(y_train,yhat)
plt.xlabel('Targets(Y Train)',fontsize=16)
plt.ylabel('Predictions(Y Hat)',fontsize=16)
plt.show()
```

Residual: Differences between targets and predictions
and Estimates for Errors

```python
#RESIDUAL PLOT
sns.displot(y_train-yhat)
plt.title('Residuals Plot',size=16)
#A much higher price is predicted than is observed
#Rarely underestimate the target
```

```python
#R Squared value
reg.score(x_train,y_train)
```

# Weights and Bias

# Testing

```python
yhat_test=reg.predict(x_test)
yhat_test
```

plt.scatter(x,y,alpha) creates a scatter plot that specifies the opacity

0.4 means 40% OPACITY

```python
plt.scatter(y_test,yhat_test,alpha=0.4)
plt.xlabel('Targets (Y_Test)',fontsize=16)
plt.ylabel('Predictions(Y_Hat)',fontsize=16)
plt.show()
```

Exponential of the log prices= Original Prices

np.exp(x) returns the exponential of x

```python
df_performance=pd.DataFrame(np.exp(yhat_test),columns=['Predictions'])
df_performance
```

```python
df_performance['Target']=np.exp(y_test)
df_performance
```

```python
df_performance.isnull().sum()
```

```python
y_test = y_test.reset_index(drop=True)
y_test
```

```python
df_performance['Target']=np.exp(y_test)
df_performance
```

```python
df_performance.isnull().sum()
```

```python
df_performance['Residual']=df_performance['Target']-df_performance['Predictions']
df_performance
```

np.absolute calulates absolute value of a specified column of data

```python
df_performance['Diff(%)']=np.absolute(df_performance['Residual']/df_performance['Target']*100)
df_performance
```

```python
df_performance.describe(include='all')
```
