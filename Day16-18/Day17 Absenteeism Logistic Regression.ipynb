```python
#Supervised Machine Learning has 0 or 1
#Kondisi: Kalau ada VALUE di column ATiH >3 maka
#Kalau TRUE >3, jadikan value 1
#Kalau FALSE <3, jadikan value 0

targets=np.where(data['Absenteeism Time in Hours']>3,1,0)
targets
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator,TransformerMixin):
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler=StandardScaler(copy,with_mean,with_std)
        self.columns=columns
        self.mean_=None
        self.var_=None
    
    def fit(self,x,y=None):
        self.scaler.fit(x[self.columns],y)
        self.mean_=np.mean(x[self.columns])
        self.var_=np.var(x[self.columns])
        return self
    
    def transform(self,x,y=None,copy=None):
        init_col_order=x.columns
        x_scaled=pd.DataFrame(self.scaler.transform(x[self.columns]),columns=self.columns)
        x_not_scaled=x.loc[:,~x.columns.isin(self.columns)]
        return pd.concat([x_not_scaled,x_scaled],axis=1)[init_col_order]
```

```python
CustomScaler
```

```python
inputs.columns.values
```

```python
column_name=[ 'Month', 'Transportation Expense','Age', 'Body Mass Index', 'Children', 'Pets']
abs_scaler=CustomScaler(column_name)
abs_scaler.fit(inputs)
```

```python
scaled_inputs=abs_scaler.transform(inputs)
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split
train_test_split(scaled_inputs,targets)
```

```python
#80-20 split
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,train_size=0.8,shuffle=True,random_state=42)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
print(reg.score(x_train,y_train))
print(reg.coef_)
print(reg.intercept_)
```

```python
predict=reg.predict(x_train)
predict
```

```python
predict==y_train
```

```python
#Number of true prediction
print(f'Accuracy:{np.sum(predict==y_train)/predict.shape[0]}')
```

```python
summary=inputs.columns.values
summary
```

```python
#Creating new dataframe table
#Transposing coefficients data -> Bcs ndarray is columns not rows
summary_table=pd.DataFrame(columns=['Features Name'],data=summary)
summary_table['Coefficients']=np.transpose(reg.coef_)
summary_table
```

```python
#Add 1 more index to the dataframe
#At row 0, give features name: intercept and value of intercept
#.sort_index() sorts table by index

summary_table.index=summary_table.index+1
summary_table.loc[0]=['Intercept',reg.intercept_[0]]
summary_table=summary_table.sort_index()
summary_table
```

```python
#Standardized Coefficients
summary_table['Odds Ratio']=np.exp(summary_table['Coefficients'])
summary_table
```

```python
#.sort_values('',ascending=False) sorts values in descending order

#Reason1: Various diseases
#Reason2: Pregnancy and giving birth
#Reason3: Poisoning
#Reason4: Light diseases
#Bias/Intercept--> Calibrates the model

summary_table.sort_values('Odds Ratio',ascending=False)
```
