```python
import numpy as np
import matplotlib.pyplot as plt
```

np.savez() save several arrays into a single file .npz format

npz files--> NumPy file type, Stores n-dimensional arrays

```python
observations=1000
x=np.random.uniform(low=-10,high=10,size=(observations,1))
z=np.random.uniform(-10,10,(observations,1))
inputs= np.column_stack((x,z))
noise=np.random.uniform(-1,1,(observations,1))
targets=2*x-3*z+5+noise
np.savez('Tensorflow Intro',inputs=inputs,targets=targets)
```

```python
train=np.load('Tensorflow Intro.npz')
input_size=2 #x,z
output_size=1 #y
```

```python
print(train)
```

tf.keras.Sequential() function that specifies how the model will be laid down

output= np.dot(inputs, weights)

tf.keras.layers.Dense (output size) takes the inputs provided to the model and calculates the dot product of the inputs and weights and biases

model.compile(optimizer,loss) configures the model for training

L2-Norm Loss: Least Sum of Squares Error

model.fit(inputs,targets) fits and trains the model

Epoch= iteration over the full dataset

verbose=0 --> Silent or no output for the training is displayed

verbose=1--> Progress Bar

verbose=2--> One line per epoch

tf.keras.optimizers.SGD(learning_rate)

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(output_size,
                                                kernel_initializer=tf.random_uniform_initializer(minval=-0.1,maxval=0.1),
                                                bias_initializer=tf.random_uniform_initializer(minval=-0.1,maxval=0.1))])
custom_optimizer=tf.keras.optimizers.SGD(learning_rate=0.002)
model.compile(optimizer=custom_optimizer,loss='mean_squared_error')
model.fit(train['inputs'],train['targets'],epochs=100,verbose=2)
```

```python
#Returns output for Weights and Bias
model.layers[0].get_weights()
```

```python
weights=model.layers[0].get_weights()[0]
print(weights)
```

```python
bias=model.layers[0].get_weights()[1]
print(bias)
```

model.predict_on_batch(data) calculates the outputs given inputs

```python
#Round ke 1 decimal place
model.predict_on_batch(train['inputs']).round(1)
```

```python
train['targets'].round(1)
```

np.squeeze removes axes of length one from an array

```python
plt.plot(np.squeeze(model.predict_on_batch(train['inputs'])),np.squeeze(train['targets']))
plt.xlabel('Predicted Output')
plt.ylabel('Real Target')
plt.show()
```
